```python
#def a fun to get both player name and stat, put them into a list
def get_stat_name_jenn(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html5lib')
    div = soup.find_all("div", class_="mod-content")[1]
    tr = div.find_all('tr', class_='oddrow')[-1]
    
    h1 = soup.find('h1')
    h1.text 
    stat_res = []
    
    for i in range(2, len(tr)):
        stat_res.append(tr.find_all('td')[i].text)
        
    
    return [h1.text] + stat_res
```
